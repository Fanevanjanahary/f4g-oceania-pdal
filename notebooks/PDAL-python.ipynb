```python
import pdal
```

```python
import numpy as np
import matplotlib.pyplot as plt
import shapely
import json
import pandas as pd

from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
%matplotlib inline
```

```python
workingfile = "../../sample-data/T_316000_235500.laz"
```

```python
pipelineJson = {
  "pipeline": [ workingfile,
    {
        "type":"filters.mortonorder",
        "reverse":"true"
    },
    {   "type":"filters.normal",
        "knn":8
    }
  ]
}
```

```python
pipeline = pdal.Pipeline(json.dumps(pipelineJson))
pipeline.validate() # check if our JSON and options were good
pipeline.loglevel = 8 #really noisy
count = pipeline.execute()
arrays = pipeline.arrays
metadata = pipeline.metadata
log = pipeline.log
```

```python
# Load Pipeline output in python objects
arr = pipeline.arrays[0]
description = arr.dtype.descr
cols = [col for col, __ in description]
df = pd.DataFrame({col: arr[col] for col in cols})
```

```python
df
```

```python
fig = plt.figure(figsize=(20,8))

ax = fig.add_subplot(111, projection='3d')
ax.set_aspect('equal')

ax.scatter(df.X[0:10000], df.Y[0:10000], df.Z[0:10000], c=df.Z[0:10000])
ax.axis('tight')
```

```python
grnd = df.where((df.Classification == 4))
```

```python
fig = plt.figure(figsize=(20,8))

ax = fig.add_subplot(111, projection='3d')
ax.set_aspect('equal')

ax.scatter(grnd.X[0:10000], grnd.Y[0:10000], grnd.Z[0:10000], c='brown')
ax.axis('tight')
```

```python
pipelineJson2 = {
  "pipeline": [ workingfile,
    {
        "type":"filters.assign",
        "assignment":"Classification[:]=0"
    },
    {
        "type":"filters.pmf",
    },
    {   "type":"filters.mortonorder",
        "reverse": True
    }
  ]
}
```

```python
%%time

pipeline = pdal.Pipeline(json.dumps(pipelineJson2))
pipeline.validate() # check if our JSON and options were good
pipeline.loglevel = 8 #really noisy
count = pipeline.execute()
arrays = pipeline.arrays
metadata = pipeline.metadata
log = pipeline.log
```

```python
arr = pipeline.arrays[0]
description = arr.dtype.descr
cols = [col for col, __ in description]
df = pd.DataFrame({col: arr[col] for col in cols})
```

```python
grnd = df.where((df.Classification == 2))
notgrnd = df.where((df.Classification != 2))
```

```python
fig = plt.figure(figsize=(20,20))

ax = fig.add_subplot(211, projection='3d')
ax.set_aspect('equal')


ax.scatter(notgrnd.X[0:20000], notgrnd.Y[0:20000], notgrnd.Z[0:20000], c=notgrnd.Z[0:20000], alpha=0.2, cmap='gist_earth')
ax.scatter(grnd.X[0:20000], grnd.Y[0:20000], grnd.Z[0:20000], c=grnd.Z[0:20000], cmap='gray')

ax2 = fig.add_subplot(212)

ax2.scatter(notgrnd.X[0:20000], notgrnd.Y[0:20000], c=notgrnd.Z[0:20000], cmap='gist_earth')
ax2.scatter(grnd.X[0:20000], grnd.Y[0:20000], c=grnd.Z[0:20000], cmap='gray')

#ax.axis('tight')
```
